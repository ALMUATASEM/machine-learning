```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from plotly.offline import iplot
import plotly.graph_objs as go
import plotly.figure_factory as ff
```

```python
timesData=pd.read_csv("/kaggle/input/world-university-rankings/timesData.csv")
```

```python
timesData.head()
```

```python
dataframe = timesData[timesData.year == 2016]
```

```python
data2016 = dataframe.loc[:,["research","citations", "teaching"]] 
```

```python
data2016["index"] = np.arange(1,len(data2016)+1) 
```

```python
data2016.head()
```

```python
fig = ff.create_scatterplotmatrix( 
                                  data2016, 
                                  diag='box',
                                  index='index', 
                                  colormap='Portland',
                                  colormap_type='cat',
                                  height=700, width=700)
iplot(fig) 
```
